### 11. Astrónomo : telescopio

```python
# --- SIMULADOR DE ASISTENTE DE TELESCOPIO (ASTRO-BOT v1.0) ---

print("Bienvenido al Centro de Control de Observación Espacial")
asistente_activo = "si"

# El ciclo while mantiene la interfaz activa hasta que el usuario decida salir
while asistente_activo == "si":
    print("\n--- MENÚ PRINCIPAL DEL TELESCOPIO ---")
    print("1. Iniciar observación de un astro")
    print("2. Consultar parámetros técnicos del equipo")
    print("3. Finalizar sesión")

    opcion = input("\nSeleccione una opción (1, 2 o 3): ")

    # Estructura if-elif-else para procesar las opciones del menú
    if opcion == "1":
        print("\n--- ASISTENTE DE OBSERVACIÓN ---")
        objetivo = input("¿Qué desea observar: 'planeta', 'luna' o 'estrella'? ").lower()

        # Simulación de IA mediante condicionales anidados
        if objetivo == "planeta":
            print("IA: Analizando coordenadas de Júpiter y Saturno...")
            distancia = input("¿El planeta está 'cerca' o 'lejos' del horizonte? ").lower()
            if distancia == "cerca":
                print("IA Sugerencia: Hay mucha turbulencia atmosférica. Use un filtro azul.")
            else:
                print("IA Sugerencia: Visibilidad óptima. Aumente la potencia del ocular.")

        elif objetivo == "luna":
            print("IA: Detectando brillo lunar...")
            fase = input("¿La luna está en fase 'llena' o 'cuarto'? ").lower()
            if fase == "llena":
                print("IA Sugerencia: El brillo es intenso. Coloque el filtro lunar de densidad neutra.")
            else:
                print("IA Sugerencia: Buen contraste en los cráteres. Use un ocular de 10mm.")

        elif objetivo == "estrella":
            print("IA: Calibrando seguimiento sideral...")
            tipo = input("¿Es una estrella 'simple' o un 'cúmulo'? ").lower()
            if tipo == "cúmulo":
                print("IA Sugerencia: Use visión periférica y un ocular de gran campo (25mm).")
            else:
                print("IA Sugerencia: Mantenga el motor de seguimiento activo para evitar el movimiento terrestre.")

        else:
            print("IA Error: Objeto no identificado en los mapas estelares.")

    elif opcion == "2":
        # Uso de operadores aritméticos para cálculos técnicos
        print("\n--- ESPECIFICACIONES TÉCNICAS ---")
        apertura = 150 # mm
        focal = 1200   # mm
        # Calculamos la relación focal (f/D)
        relacion_focal = focal / apertura

        print(f"Apertura del lente: {apertura} mm")
        print(f"Distancia focal: {focal} mm")
        print(f"Relación focal calculada: f/{relacion_focal}")

    elif opcion == "3":
        print("\nIA: Cerrando protectores de los lentes... ¡Cielos despejados, Astrónomo!")
        # Cambio de la variable para terminar el ciclo while
        asistente_activo = "no"

    else:
        print("\nError: Entrada no reconocida por el sistema.")

print("\n--- SISTEMA APAGADO ---")
print("Programa elaborado por Chavez Iris NL = 11")
```
